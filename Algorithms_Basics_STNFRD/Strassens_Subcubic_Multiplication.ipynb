Problem statement: Multiplying two matrices:
- The elements in matrices are just entries that we can add and multiply.
so, you take two n x n matrices, x and y, and multiply them producing a new n x n matrix, Z?

Well, recall that the IJ entry of Z that means the entry in the Ith row and Jth  column, is simply the dot product of the Ith row of X with the Jth column of Y. 



So ultimately, the ZIJ entry boils down to a sum over N things, where each of the constituent products is just the xIK entry. 

Dot product: product of individual components and add up the results.

Note: One thing to note is where we often use N to denote the input size, here we're using N to note the dimension of each of these matrices.
- the input size is not N. The input size is quite a bit bigger than N.
- Specifically, each of these are N by N matrices and contain N squred entries. So since presumably we have to read the input which has size N squred. Which happens to produce the output that also has size N squared.
- the best we can really hoe for running time of an multiplication of two matrices algorithm is the running time of N squared.
