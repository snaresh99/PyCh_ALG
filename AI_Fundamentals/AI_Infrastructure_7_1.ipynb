# Data Center Platforms.

* Full stack platforms for accelerated computing.
- accelerated computing services, software and systems for the modern data center.
- Modern data centers are the key in solving some of the world's most important scientific, industrial, and big data challenges using high performance computing and AI.
- Accelerated computing is often referred to as a full stack challenge because it involves optimizing and integrating various components across multiple layers of the technology stack to achieve optimal performance for specialized workloads.
- Furthermore, accelerated computing represents a data center scale issue.
- As the modern data center essentially serves as the computer.
- Applications span the entire data center, making it crucial to optimize all the diverse components within it.
- at the foundation are the hardware technologies, GPUs, CPUs, and DPU's that form the basis for building servers.
- Sitting at top of these servers is a software stack encompassing CUDA and DOCA, the proramming models for GPUs and DPUs respectively, 
along with numerous software libraries that transparently provides acceleration to developers across different hardware products such as 
CUDA X for GPU acceleration.
