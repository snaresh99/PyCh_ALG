



Objective: Brute Force Method:
We want to find two positoins that would maximize our area. How about we search every possible area?





















```python
# Finding max area with given heights
# heights = [5,9,2,1,4]
"""
The time complexity for thisis actually big O of n squared and this is the result of 
n multiplied with n.

n for the outer loop and n multiplied by the n for the inner loop.

"""

def maxArea(heights):
    max_area = 0
    n = len(heights)
    for p1 in range(0, n):
        for p2 in range(p1, n):
            length = min(heights[p1], heights[p2])
            width = p2 - p1
            area = length * width
            max_area = max(max_area, area)

    return max_area
```

```python
# Find the max area test case 1
heights = [5,9,2,1,4]
water_container = maxArea(heights)
print(water_container)

# Find the max area test case 12
heights2 = [5,9,2,1,4,5]
water_container2 = maxArea(heights2)
print(water_container2)
```

Better than bruteforce method: Approach to reduce the time complexity from Big O of n squred.



```python
# Efficient way to calculaate the area of water container

def WaterArea(heights):
    l = 0
    r = len(heights)-1
    max_area = 0
    print("length of heights", len(heights))
    while(l < r):
        width = r-l
        min_height = min(heights[l], heights[r])
        print("Minimum heights", min_height)
        area = min_height * width
        max_area = max(max_area, area)
        if(heights[l] < heights[r]):
            l += 1
        else:
            r -= 1

    return max_area
```

```python
# Test to verify the above defined efficient way

heights = [1,8,6,2,5,7]
eff_area = WaterArea(heights)
print(eff_area)
```
