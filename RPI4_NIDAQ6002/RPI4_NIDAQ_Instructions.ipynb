import time

gpio_pin = 17

try:

        GPIO.output(gpio_pin, GPIO.HIGH)  # Set pin high
        print("GPIO Pin HIGH")
        time.sleep(1)                      # Wait for 1 second
        GPIO.output(gpio_pin, GPIO.LOW)   # Set pin low

    pass
finally:

```python
import nidaqmx
from nidaqmx.constants import AcquisitionType

# Configure the task
with nidaqmx.Task() as task:
    task.ai_channels.add_ai_voltage_chan("Dev1/ai0")  # Change to your DAQ channel
    task.timing.cfg_samp_clk_timing(1000.0, sample_mode=AcquisitionType.CONTINUOUS)

    # Start the task
    task.start()

    try:
        while True:
            data = task.read(number_of_samples_per_channel=1)
            print(f"DAQ Reading: {data}")
    except KeyboardInterrupt:
        pass
```
