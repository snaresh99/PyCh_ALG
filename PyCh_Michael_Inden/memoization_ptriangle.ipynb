# Memoization for Pascal's triangle:
Pascal triangle are defined recursively, as are the fibonacci numbers.



```python
# Pascal triangle using recursion

def pascal_rec(row, col):
    # recursive termination: top

    if col == 1 and row == 1:
        return 1
    # Recursive termination: Borders

    if col == 1 or col == row:
        return 1
    
    # recursive descent
    print(pascal_rec(row-1, col) + pascal_rec(row-1, col -1))
    return pascal_rec(row-1, col) + pascal_rec(row-1, col -1)

if __name__ == "__main__":
    a = pascal_rec(3,3)
    print(a)

```
